```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Import dataset
ramen = pd.read_csv('data/ramen-ratings.csv')
```

```python
# Look at the data and some initial statistical analysis
ramen.head()
```

```python
ramen.info()
```

```python
print(ramen.describe())
print(ramen.describe(include='O'))
```

```python
ramen['Style'].unique()
# There are a couple of null values in this column
```

```python
null_styles = ramen[ramen['Style'].isna()]
print(null_styles)
# There's only two of these, so I think it's fine to just drop them
ramen_clean = ramen.dropna(subset=['Style'])
print(ramen_clean.shape)
```

```python
ramen.isna().sum()
```

```python
# What is the Top Ten Column?
ramen['Top Ten'].unique()
```

#### Initial Observations
- There's only one numerical column, and that is the review number
- Other category descriptions:

`Brand`: brand of ramen

`Variety`: More or less the flavor of the ramen

`Style`: How the ramen was packaged/served

`Country`: Country of origin

`Stars`: The actual rating of the review; for some reason it's a string.

Cleaning and investigating time

```python
print(ramen_clean['Stars'].unique())
```

```python
# Problem. There's an 'Unrated' rating
UNRATED_MASK = (ramen_clean['Stars'] == 'Unrated')
print(len(ramen_clean[UNRATED_MASK]))
```

```python
# Considering there's only three, I think it's fine to just drop these too
ramen_clean = ramen_clean[~UNRATED_MASK]
ramen_clean['Stars'].unique()
```

```python
ramen_clean['Stars'] = ramen_clean['Stars'].astype(float)
ramen_clean['Stars'].round(2)
```

```python
# I don't want to deal with the top ten column right now, so I'm going to move on to descriptive questions
## What is the distribution of Ramen Brands? Maybe just stick to the top 10 or 20 or something
ramen_clean['Brand'].value_counts()[:15]
```

```python
## Are there any ramens with a 5 star rating?
FIVE_STARS = (ramen_clean['Stars'] == 5.0)
five_star_ramen = ramen_clean[FIVE_STARS]
print(five_star_ramen)
```

```python
## What is the distribution of countries that these 5 star ramens come from?
five_star_ramen['Country'].value_counts(normalize=True)
```

```python
## What are the average ratings of the different styles of ramen?
ramen_clean.groupby('Style')['Stars'].mean()
```

```python
# Wow, bar has a perfect 5 star rating, and box is pretty high up there too
ramen_clean['Style'].value_counts()
```

It looks like that's because there's only a few instances of each one...

```python
# I guess I can try dealing with the Top Ten column now

# First I'll just fill the missing values with a not-applicable string
ramen_clean['Top Ten'].fillna('N/A', inplace=True)

# There are also some newline characters in there, so I'll replace those too
ramen_clean['Top Ten'] = ramen_clean['Top Ten'].str.replace('\n', 'N/A')
## I learned that Series.str.replace() doesn't do it in place

# Time to see if the null values are gone
print(ramen_clean.isna().sum())
print(ramen_clean['Top Ten'].unique())
```

```python
# Let's look at all the ramen where the Top Ten column is applicable
TOP_TEN = (ramen_clean['Top Ten'] != 'N/A')
top_ten_ramen = ramen_clean[TOP_TEN]
top_ten_ramen
```

```python
# How about all the #1s?
NUMBER_ONES = (top_ten_ramen['Top Ten'].str.contains('1$'))
top_ten_ramen[NUMBER_ONES]
```

```python
# A plot for fun
fig, ax = plt.subplots(figsize=(10, 8))
sns.histplot(data=ramen_clean, x='Style');
```
